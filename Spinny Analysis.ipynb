```python
# As we have null values in many columns, so we will fix it.
# 1. For numerical columns, fill with mean
numeric_columns = ['year', 'odometer']
for column in numeric_columns:
    data[column] = data[column].fillna(data[column].mean())

# 2. For non-numerical columns with null values greter than topmost occuring type.
for column in ['model', 'VIN', 'condition', 'cylinders', 'drive', 'size', 'type', 'paint_color', 'image_url']:
    data[column] = data[column].fillna('Unknown')

# 3. For non-numerical columns with fewer null values.
for column in ['manufacturer', 'fuel', 'title_status', 'transmission']:
    data[column] = data[column].fillna(data[column].mode()[0])

# 4. Convert 'posting_date' to datetime objects to handle errors while filling null values.
data['posting_date'] = pd.to_datetime(data['posting_date'], errors = 'coerce', utc = True)
data['posting_date'] = data['posting_date'].fillna(data['posting_date'].mean())
```

```python
data.isna().sum()
```

```python
# Converted float datatype of year to integer datatype
data['year'] = data['year'].astype(int)

# Replaced cyclinder word from that column as the column name is already cylinder
data['cylinders'] = data['cylinders'].str.replace('cylinders', "")

# Dropping image_url and VIN as its of no use in this particular analysis.
data = data.drop('image_url', axis = 1)
data = data.drop('VIN', axis = 1)
```

```python
# IQR Method
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)

# outliers
data.loc[(data['price'] > upper_limit) | (data['price']< lower_limit)]
```
